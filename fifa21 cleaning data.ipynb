```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('fifa21 raw data.csv', low_memory=False)
new_df = pd.DataFrame([])
pd.set_option('display.max_columns',None)
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
# extract 'FullName' without additional parameters(eg. characters like 'é') from 'playerUrl' in string format
new_df['FullName'] = df['playerUrl'].apply(lambda text: text.split('/')[5:-2][0].replace('-', ' ').title())
new_df.head(10)
```

```python
new_df['Club'] = df['Club'].apply(lambda text: text.strip() if type(text)==str else text)
new_df.head(10)
```

```python
# filter and convert all 'Height' values to centimeters and 'Weight' values to kilograms in integer format
def method2(text):
    if 'cm' in text:
        text = int(text.replace('cm', ''))
    elif "'" in text and '"' in text:
        feet,inch = text.split("'")
        feet = int(feet)
        inch = int(inch.replace('"', ''))
        text = int(((feet*12) + inch) * 2.54)
    return text

new_df['Height(cm)'] = df['Height'].apply(method2)

def method3(text):
    if 'kg' in text:
        text = int(text.replace('kg', ''))
    elif 'lbs' in text:
        pound = int(text.replace('lbs', ''))
        text = int(pound * 0.45359237)
    return text

new_df['Weight(kg)'] = df['Weight'].apply(method3)
new_df.head(10)
```

```python
# convert 'Joined' column type to datetime format
new_df['Joined'] = df['Joined'].apply(pd.to_datetime)
new_df.head(10)
```

```python
# filter and extract 'ContractStart' and 'ContractEnd' from 'Joined', 'Loan Date End' and 'Contract' in string format
new_df[['contractmonth','ContractEnd']] = df['Loan Date End'].str.split(',', expand = True).apply(lambda x: x.str.strip())
new_df['ContractStart'] = df['Joined'].apply(lambda x : x[-4:])
new_df['ContractEnd'] = new_df['ContractEnd'].fillna(df['Contract'].apply(lambda x : x[-4:]))
new_df.drop(['contractmonth'], axis=1, inplace=True)
new_df.head(10)
```

```python
# extract and fill values in 'LoanStatus' from 'Loan Date End' in string format
new_df['LoanStatus'] = df['Loan Date End'].fillna("No").apply(lambda text: "Yes" if len(text)>4 else "No")
new_df.head(10)
```

```python
# extract and fill values in 'ContractStatus' from 'Contract' in string format
def method1(text):
    for i in text.split():
        if "Free" in i:
            return "Free"
        elif "Loan" in i:
            return "Loan"
        elif "~" in i:
            return "Contract"
        
new_df['ContractStatus'] = df['Contract'].apply(method1)
new_df.head(10)
```

```python
# filter and convert all 'Value','Wage' and 'ReleaseClause' values to integer format
def method5(texts):
    game = []
    for text in texts:
        if 'K' in text:
            text = text.replace('K','').replace('€','')
            text = float(text) * 1000
            text = int(text)
            game.append(text)
        elif 'M' in str(text):
            text = text.replace('M','').replace('€','')
            text = float(text) * 1000000
            text = int(text)
            game.append(text)
        else:
            text = int(text.replace('€',''))
            game.append(text)
    return game

new_df[['Value(€)','Wage(€)','ReleaseClause(€)']] = df[['Value','Wage','Release Clause']].apply(method5)
new_df.head(10)
```

```python
# filter and convert all 'WeakFoot','SkillMove' and 'InjuryResistance' values to integer format
new_df[['WeakFoot(★)','SkillMove(★)','InjuryResistance(★)']] = df[['W/F','SM','IR']].apply(lambda text: [int(i.replace('★', '')) for i in text])
new_df.head(10)
```

```python
# filter and convert all 'Hits' values to integer format
new_df['Hits'] = df['Hits'].fillna('0').apply(lambda text: int(float(text.replace('K','')) * 1000) if 'K' in str(text) else int(text))
new_df.head(10)
```

```python
# add remaining columns to new_df
new_df[['PhotoUrl','Nationality','Age','Overall','Potential','Positions','PreferredFoot','BestOverallRating','BestPosition','Attacking','Crossing','Finishing','HeadingAccuracy','ShortPassing','Volleys','Skill','Dribbling','Curve','FKAccuracy','LongPassing','BallControl','Movement','Acceleration','SprintSpeed','Agility','Reactions','Balance','Power','ShotPower','Jumping','Stamina','Strength','LongShots','Mentality','Aggression','Interceptions','Positioning','Vision','Penalties','Composure','Defending','Marking','StandingTackle','SlidingTackle','Goalkeeping','GKDiving','GKHandling','GKKicking','GKPositioning','GKReflexes','TotalStats','BaseStats','AttackingWorkrate','DefendingWorkrate','Pace','Shoot','Pass','Dribble','Defense','Physical']] = df[['photoUrl','Nationality','Age','↓OVA','POT','Positions','Preferred Foot','BOV','Best Position','Attacking','Crossing','Finishing','Heading Accuracy','Short Passing','Volleys','Skill','Dribbling','Curve','FK Accuracy','Long Passing','Ball Control','Movement','Acceleration','Sprint Speed','Agility','Reactions','Balance','Power','Shot Power','Jumping','Stamina','Strength','Long Shots','Mentality','Aggression','Interceptions','Positioning','Vision','Penalties','Composure','Defending','Marking','Standing Tackle','Sliding Tackle','Goalkeeping','GK Diving','GK Handling','GK Kicking','GK Positioning','GK Reflexes','Total Stats','Base Stats','A/W','D/W','PAC','SHO','PAS','DRI','DEF','PHY']]
new_df.head(10)
```

```python
# rearrange columns in original order to ensure new_df is as similar as possible to df
new_df = new_df.reindex(columns=['FullName', 'PhotoUrl', 'Nationality', 'Age', 'Overall', 'Potential', 'Club', 'ContractStart', 'ContractEnd', 'ContractStatus', 'Positions', 'Height(cm)', 'Weight(kg)', 'PreferredFoot', 'BestOverallRating', 'BestPosition', 'Joined', 'LoanStatus', 'Value(€)','Wage(€)','ReleaseClause(€)', 'Attacking', 'Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Skill', 'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Movement', 'Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'Power', 'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots', 'Mentality', 'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure', 'Defending', 'Marking', 'StandingTackle', 'SlidingTackle', 'Goalkeeping', 'GKDiving', 'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes', 'TotalStats', 'BaseStats', 'WeakFoot(★)', 'SkillMove(★)', 'AttackingWorkrate', 'DefendingWorkrate', 'InjuryResistance(★)', 'Pace', 'Shoot', 'Pass', 'Dribble', 'Defense', 'Physical', 'Hits'])
new_df.head(10)
```
